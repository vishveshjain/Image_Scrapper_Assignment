# USING Youtube DATA API -

```python
import pandas as pd
from googleapiclient.discovery import build

api_key=""  # PLEASE INSERT YOUR OWN API KEY

PW_Skill_channel_id = 'UCphU2bAGmw304CFAzy0Enuw'

youtube = build('youtube', 'v3', developerKey=api_key)


def extract_video_url(youtube, channel_id):
    request = youtube.channels().list(
        part="snippet,contentDetails,statistics",
        id=channel_id
    )
    response = request.execute()


    return response

data=extract_video_url(youtube, PW_Skill_channel_id)
data
```

```python
playlist=data['items'][0]['contentDetails']['relatedPlaylists']['uploads']
playlist
```

# Etracting Video ids 

```python
def get_video_details(youtube,playlist,pg_token=''):
    request = youtube.playlistItems().list(
        part="contentDetails",
        playlistId=playlist,
        maxResults=50
    )
    response = request.execute()
    list_data=response['items']
    video_ids=[x['contentDetails']['videoId'] for x in list_data]
    return video_ids
    

video_ids=get_video_details(youtube,playlist)
video_ids
```

# extracting video details

```python
def video_details(youtube, video_ids):
    request = youtube.videos().list(
    part="snippet,contentDetails,statistics",
    id=','.join(video_ids)
    )
    response = request.execute()

    list_data = response['items']

    data= {
    'thumbnail_url' : [x['snippet']['thumbnails']['default']['url'] for x in list_data],
    'title' : [x['snippet']['title'] for x in list_data],
    'publishedAt' : [x['snippet']['publishedAt'] for x in list_data],
    'view_count' : [x['statistics']['viewCount'] for x in list_data]
    }

    return data


all_data=video_details(youtube, video_ids)
all_data
```

# Converting to DataFrame

```python
df=pd.DataFrame(all_data)
df['video_url'] = 'https://www.youtube.com/watch?v='+pd.Series(video_ids)
df
```
